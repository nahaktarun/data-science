# Q-Q Plot

```python
import numpy as np
import pylab
import scipy.stats as stats

#N(0,1)
std_normal = np.random.normal(loc=0,scale = 1,size=1000) #generate random  observation

# 0 to 100th percentiles of std-normal

for i in range(0,101):
    print(i, np.percentile(std_normal,i))
```

```python
#generate 100 samples from N(20,5)
measurements = np.random.normal(loc=20,scale=5,size=50)


stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()
```

```python
# generate 100 samples from N(20,5)
measurement = np.random.uniform(low=-1,high=1,size=10000)


stats.probplot(measurement, dist ="norm",plot=pylab)
pylab.show()
```
