# Pyplot 

**matplotlib.pyplot** is a collection of command style functions that make matplotlib work like MATLAB

Each pyplot function makes some change to a figure: i.e create a figure, create a plotting area in a figure, plots some lines in a plotting area, decortates the plot with labels etc.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel("numbers")
plt.xlabel("Indices")
plt.title("Myplot")
plt.show()
```

### plot x vs y

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.ylabel("squares")
plt.xlabel("numbers")
plt.title("Square of values")
plt.grid() #grid on

plt.show()
```

for every x,y pair of argument, there is an optional third argument which is the format string that indicates the color and line type of the plot

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') #r-> red o-> o shape dot
plt.grid()
plt.show()
```

```python
import numpy as np

t = np.arange(0.,5.,0.2)
#blue dashes, red squares and green triangles
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.5,'g^',label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

# Controlling line properties

### use key

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, linewidth=5.0)
plt.show()
```

### set properties

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]
lines = plt.plot(x1, y1, x2, y2)

# use keyword args
plt.setp(lines[0], color = 'r', linewidth = 2.0)

# matlab style string value pair
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)
plt.grid()
```

## Working with multiple figures and axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)

plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1, 2, 3])
plt.subplot(212)
plt.plot([4, 5, 6])

plt.figure(2)
plt.plot([4, 5, 6])

plt.figure(1)
plt.subplot(211)
plt.title('Easy as 1, 2, 3')
plt.show()
```
